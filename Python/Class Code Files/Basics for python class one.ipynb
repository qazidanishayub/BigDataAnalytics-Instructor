# **Homework**
Create a function that checks if a number entered is a prime number or not

```python
def primeNumberCheck():
  A = int(input('Enter a number: '))
  check = 0
  if A>= 0:
    for i in range(2, A+1):
        if (A % i) == 0:
          check+=1
        else:
          continue
  if check>1:
    print('its not a prime number')
  else:
    print('its a prime number')
```

```python
primeNumberCheck()
```
