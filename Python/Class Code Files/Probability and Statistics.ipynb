Measures of central tendency are statistical measures that provide a single value that represents the center or typical value of a dataset. The three main measures of central tendency are mean, mode, and median:

Mean: The mean, also known as the average, is the sum of all values in a dataset divided by the total number of values.

```python
from collections import Counter

def calculate_mode(data):
    # Count occurrences of each value
    count = Counter(data)
    # Find the most common value(s)
    mode_values = count.most_common(1)
    return mode_values[0][0]

# Example data
data = [10, 20, 30, 30, 40, 40, 40, 50]

# Calculate mode
mode = calculate_mode(data)
print("Mode:", mode)
```

Mode: The mode is the value that appears most frequently in a dataset. A dataset may have one mode (unimodal), multiple modes (multimodal), or no mode if all values occur with the same frequency.

```python
def calculate_median(data):
    sorted_data = sorted(data)
    n = len(data)
    if n % 2 == 0:
        # If even number of values, average the two middle values
        mid = n // 2
        median = (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        # If odd number of values, take the middle value
        median = sorted_data[n // 2]
    return median

# Example data
data = [10, 20, 30, 40, 50]

# Calculate median
median = calculate_median(data)
print("Median:", median)
```

```python
def calculate_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        # If even number of values, average the two middle values
        mid = n // 2
        median = (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        # If odd number of values, take the middle value
        median = sorted_data[n // 2]
    return median

# Example data
data = [10, 20, 30, 40, 50]

# Calculate median
median = calculate_median(data)
print("Median:", median)
```

```python
import numpy as np

def calculate_variance(data):
    mean = np.mean(data)
    variance = np.mean((data - mean) ** 2)
    return variance

def calculate_standard_deviation(data):
    variance = calculate_variance(data)
    std_dev = np.sqrt(variance)
    return std_dev

# Example data
data = [10, 20, 30, 40, 50]

# Calculate variance and standard deviation
variance = calculate_variance(data)
std_deviation = calculate_standard_deviation(data)

print("Variance:", variance)
print("Standard Deviation:", std_deviation)
```

Measures of dispersion provide information about the spread or variability of a dataset. Two common measures of dispersion are variance and standard deviation. Additionally, coefficient of variation, skewness, and kurtosis are statistical measures that provide insights into the shape and distribution of a dataset.

# Variance:
 Variance measures how much the values in a dataset deviate from the mean. It is calculated as the average of the squared differences between each value and the mean.

# Standard Deviation:
 Standard deviation is the square root of the variance. It represents the average distance of each data point from the mean and provides a measure of the spread of the data.

```python
def calculate_coefficient_of_variation(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    coefficient_of_variation = (std_dev / mean) * 100
    return coefficient_of_variation

# Example data
data = [10, 20, 30, 40, 50]

# Calculate coefficient of variation
cv = calculate_coefficient_of_variation(data)
print("Coefficient of Variation:", cv)
```

Coefficient of Variation: Coefficient of variation (CV) is a relative measure of dispersion that compares the standard deviation to the mean. It is expressed as a percentage and is useful for comparing the variability of datasets with different units or scales.

```python
from scipy.stats import skew

# Example data
data = [10, 20, 30, 40, 50]

# Calculate skewness
skewness = skew(data)
print("Skewness:", skewness)
```

# Skewness:
Skewness measures the asymmetry of the distribution of data around its mean. A positive skew indicates that the data is skewed to the right, meaning that the tail on the right side of the distribution is longer. A negative skew indicates that the data is skewed to the left.

```python
from scipy.stats import kurtosis

# Example data
data = [10, 20, 30, 40, 50]

# Calculate kurtosis
kurt = kurtosis(data)
print("Kurtosis:", kurt)
```

# Kurtosis:
 Kurtosis measures the peakedness or flatness of the distribution of data. It compares the tails of the distribution to the tails of a normal distribution. Positive kurtosis indicates a sharper peak (leptokurtic), while negative kurtosis indicates a flatter peak (platykurtic).

**Measures of Dispersion:**

**a. Variance:** Variance measures how much the values in a dataset vary from the mean. It is the average of the squared differences from the mean.

```python
def calculate_variance(data):
    mean = sum(data) / len(data)
    squared_diff = [(x - mean) ** 2 for x in data]
    variance = sum(squared_diff) / len(data)
    return variance

# Example data
data = [10, 20, 30, 40, 50]

# Calculate variance
variance = calculate_variance(data)
print("Variance:", variance)
```

# Standard Deviation:
 Standard deviation is the square root of the variance and measures the average deviation of the values from the mean.

```python
import math

def calculate_std_dev(data):
    variance = calculate_variance(data)
    std_dev = math.sqrt(variance)
    return std_dev

# Example data
data = [10, 20, 30, 40, 50]

# Calculate standard deviation
std_dev = calculate_std_dev(data)
print("Standard Deviation:", std_dev)
```
